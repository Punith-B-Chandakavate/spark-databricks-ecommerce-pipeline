## Create the Ecommerce Catalog
Create a Unity Catalog named `ecommerce` with a managed storage location.create a ecommerce catalog

```python
%sql

CREATE CATALOG IF NOT EXISTS ecommerce
MANAGED LOCATION 'abfss://uc-data@stgpunithadlsdevus001.dfs.core.windows.net/ecommerce_catalog'
COMMENT 'Ecommerce project catalog in east us, backed by external location';
```

## Switch to the Ecommerce Catalog
Set the active catalog to `ecommerce`.

```python
%sql

USE CATALOG ecommerce
```

## Create the Bronze, Silver, and Gold Schemas
Create the standard medallion architecture schemas for the project.

```python
%sql

CREATE SCHEMA IF NOT EXISTS ecommerce.bronze;
CREATE SCHEMA IF NOT EXISTS ecommerce.silver;
CREATE SCHEMA IF NOT EXISTS ecommerce.gold;
```
